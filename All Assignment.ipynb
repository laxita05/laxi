## To compare residual plots

```python
import pandas as pd
import numpy as np

# Sample DataFrame
data = pd.DataFrame({
    'Floor': [1.0, 1.5, 2.0, 2.5, 3.0, 3.5],
    'Sale_Price': [431826.297032, 542188.988482, 599260.70846, 798774.47205, 551141.645425, 662634.375]
})

# Calculate the overall mean of Sale_Price
overall_mean = data['Sale_Price'].mean()

# Create a pivot table to calculate the mean of 'Sale_Price' for each 'Floor'
pivot_table = pd.pivot_table(data, values='Sale_Price', index='Floor', aggfunc='mean')

# Apply the floor function to the mean values
floor_mean = pivot_table.apply(np.floor)

# Initialize the 'floor_mean' column with default value
data['floor_mean'] = 0

# Fill the 'floor_mean' column with the floored mean prices
for floor_value in floor_mean.index:
    data.loc[data['Floor'] == floor_value, 'floor_mean'] = floor_mean.loc[floor_value, 'Sale_Price']

# Calculate mean_difference (difference from the overall mean)
mean_difference = data['Sale_Price'] - overall_mean

# Calculate floor_mean_difference (difference from the floored mean price)
floor_mean_difference = data['Sale_Price'] - data['floor_mean']

# Display sizes and values
print("mean_difference size:", mean_difference.size)
print("floor_mean_difference size:", floor_mean_difference.size)
print("\nmean_difference:\n", mean_difference)
print("\nfloor_mean_difference:\n", floor_mean_difference)
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame
data = pd.DataFrame({
    'Floor': [1.0, 1.5, 2.0, 2.5, 3.0, 3.5],
    'Sale_Price': [431826.297032, 542188.988482, 599260.70846, 798774.47205, 551141.645425, 662634.375]
})

# Calculate the overall mean of Sale_Price
overall_mean = data['Sale_Price'].mean()

# Create a pivot table to calculate the mean of 'Sale_Price' for each 'Floor'
pivot_table = pd.pivot_table(data, values='Sale_Price', index='Floor', aggfunc='mean')

# Apply the floor function to the mean values
floor_mean = pivot_table.apply(np.floor)

# Initialize the 'floor_mean' column with default value
data['floor_mean'] = 0

# Fill the 'floor_mean' column with the floored mean prices
for floor_value in floor_mean.index:
    data.loc[data['Floor'] == floor_value, 'floor_mean'] = floor_mean.loc[floor_value, 'Sale_Price']

# Calculate MSE1: Mean Squared Error for the overall mean model
MSE1 = ((data['Sale_Price'] - overall_mean) ** 2).mean()

# Calculate MSE2: Mean Squared Error for the floor mean model
MSE2 = ((data['Sale_Price'] - data['floor_mean']) ** 2).mean()

# Calculate R^2 value
R2 = 1 - (MSE2 / MSE1)

# Output results
print("MSE1 (Mean Squared Error for Overall Mean Model):", MSE1)
print("MSE2 (Mean Squared Error for Floor Mean Model):", MSE2)
print("R^2 Value:", R2)
```
